```python
from tkinter import *
expr=""
def clr():
     global expr
     expr=""
     equation.set("")

def number(num):
     global expr
     expr=expr+str(num)
     equation.set(expr)

def equalpress():
    try:
         global expr
         res= str(eval(expr))
         equation.set(res)
         expr=""
    except:
        equation.set("error")
        expr=""

root=Tk()

root.configure(bg="black")

root.title("caculator")

equation=StringVar()

root.geometry("300x300")
expr_field=Entry(root,textvariable=equation)
expr_field.grid(columnspan=4,ipadx=70,ipady=20)
button0=Button(root,text='0', command=lambda:number(0),height=1,width=7)
button0.grid(row=5,column=1)
button1=Button(root,text='1',command=lambda:number(1),height=1,width=7)
button1.grid(row=2,column=0)
button2=Button(root,text='2',command=lambda:number(2),height=1,width=7)
button2.grid(row=2,column=1)
button3=Button(root,text='3',command=lambda:number(3),height=1,width=7)
button3.grid(row=2,column=2)
button4=Button(root,text='4',command=lambda:number(4),height=1,width=7)
button4.grid(row=3,column=0)
button5=Button(root,text='5',command=lambda:number(5),height=1, width=7)
button5.grid(row=3,column=1)
button6=Button(root,text=6,command=lambda:number(6),height=1,width=7)
button6.grid(row=3,column=2)
button7=Button(root,text='7',command=lambda:number(7), height=1,width=7)
button7.grid(row=4,column=0)
button8=Button(root,text='8',command=lambda:number(8),height=1,width=7)
button8.grid(row=4,column=1)
button9=Button(root,text='9',command=lambda:number(9),height=1, width=7)
button9.grid(row=4,column=2)
p=Button(root,text='+',command=lambda:number("+"),height=1 ,width=7)
p.grid(row=2,column=3)
m=Button(root,text='-',command=lambda:number("-"),height=1,width=7)
m.grid(row=3,column=3)
d=Button(root,text='/',command=lambda:number("/"),height=1,width=7)
d.grid(row=4,column=3)
multi=Button(root,text='',command=lambda:number(""),height=1,width=7)
multi.grid(row=5,column=3)
e=Button(root,text='=',command=lambda:equalpress(),height=1, width=7) 
e.grid(row=5,column=0)
c=Button(root,text='clear',command=clr,height=1,width=7)
c.grid(row=5,column=2)
root.mainloop()
```
